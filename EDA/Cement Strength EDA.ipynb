# Cement Strength EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns
sns.set()
import warnings 
warnings.filterwarnings("ignore")
```

data set from : https://archive.ics.uci.edu/ml/datasets/Concrete+Compressive+Strength

```python

data = pd.read_excel("https://archive.ics.uci.edu/ml/machine-learning-databases/concrete/compressive/Concrete_Data.xls")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
# plot the data 
plot.figure( figsize = (15, 15), facecolor='white')
plotnumber = 1

for col in data.columns:
    ax = plot.subplot(4,3, plotnumber)
    sns.distplot(data[col])
    plot.xlabel(col, fontsize=14)
    plotnumber+=1
plot.tight_layout()
```

Clearly, there is skewness in the data.

### Data Transformation

```python
data.head()
```

```python
X = data.drop(['Concrete compressive strength(MPa, megapascals) '], axis= 1, inplace = False)
Y = data[data.columns[-1]]
```

```python
# log transformation 
# adding 1 to each value in dataset, since there are zeros in dataset

for col in X.columns:
    X[col]+=1
    X[col] = np.log(X[col])
```

```python
X.head()
```

```python
plot.figure(figsize = (15,15), facecolor = 'white')
plotnumber = 1
for col in X.columns:
    ax = plot.subplot(3,3, plotnumber)
    sns.distplot(X[col])
    plot.xlabel(col, fontsize = 12)
    plotnumber+=1
plot.show()
```

```python
# check for the outliers
plot.figure(figsize = (15, 15), facecolor = 'white')
plotnumber = 1

for col in X.columns:
    ax = plot.subplot(4,3, plotnumber)
    sns.boxplot(X[col])
    plot.xlabel(col, fontsize = 10)
    plotnumber += 1
plot.show()
```

Outliers are affordable we dont need to remove 2 or 3 records for the sake of outliers in few columns

```python
# check how features are related to target column
plot.figure(figsize = (20,20), facecolor = 'white')
plotnumber = 1

for col in X.columns:
    ax = plot.subplot(4,3,plotnumber)
    sns.scatterplot(X[col], Y)
    plot.xlabel(col, fontsize = 10)
    plotnumber += 1
plot.show()
```

The relationship doesn't quite look like linear but we will implement linear regression and random forest on our data. We will compare our results from both the models

```python
# checking for the correlation in data
plot.figure(figsize = (10,10))

sns.heatmap(X.corr().abs(), vmin = -0.5, vmax = 1, annot = True)
```

GREAT, none of our column seem to be correlated.
